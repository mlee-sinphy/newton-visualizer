```python
from newton_visualizer.parser import parse_function_and_derivative
from newton_visualizer.solver import newton_raphson

from newton_visualizer.plotting import (
    plot_function_with_iterations,
    plot_newton_iterations_with_tangents,
    plot_residual_convergence,
)
```

```python
expression = "x**2 - 2"
x0 = 1.0

f, df = parse_function_and_derivative(expression)

result = newton_raphson(
    f=f,
    df=df,
    x0=x0,
)

result
```

## Numerical Result

The Newton-Raphson method converged in 4 iterations.

The final approximation is close to the positive root of \(x^2 - 2 = 0\), namely \(\sqrt{2}\).

The residual is approximately \(4.51 \times 10^{-12}\), which means that the computed value satisfies \(f(x_n) \approx 0\) with high numerical accuracy.

```python
plot_function_with_iterations(
    f=f,
    result=result,
    x_min=0,
    x_max=2,
)
```
